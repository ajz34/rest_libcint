## test_ECPscalar

```python
out = mol.intor("ECPscalar")
out_c = out.transpose(1, 0)
assert out_c.flags.c_contiguous
scale = np.linspace(-1, 1, out.size)
print(out_c.sum())
print((out_c.flatten() * scale).sum())
```
